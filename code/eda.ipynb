```python
import pickle

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from randomfunctions import print_cols

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
base_data = pickle.load(open('../assets/pickles/base_data.pickle', 'rb'))
base_data.set_index('id', inplace=True)
```

```python
def artist_name(row):
    return [artist['name'] for artist in row['artists']]

def artist_id(row):
    return [artist['id'] for artist in row['artists']]
```

```python
base_data['artist_names'] = base_data.apply(artist_name, axis=1)
base_data['artist_ids'] = base_data.apply(artist_id, axis=1)

base_data.drop('artists', axis=1, inplace=True)
```

```python
cols_to_remove = ['analysis_url', 'uri', 'disc_number', 'external_ids', 'external_urls', 
                  'type', 'preview_url']

base_data.drop(cols_to_remove, axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.distplot(base_data['popularity'], ax=ax)
plt.title('I like either obscure or popular songs', fontsize=20);
```
